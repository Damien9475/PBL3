Un algorithme est une suite d'opérations qui résolvent un problème. Il est important, dans un souci de performance, d'évaluer la « qualité » d'un algorithme. Ce souci de performance peut être compris de deux manières différentes : soit en **temps d'exécution du programme**, soit en termes de stockage. 

<div style='color: cadetblue'>
    <h4><u>Problème 1</u> :  Rechercher un mot dans un dictionnaire contenant <em>n</em> mots</h4>
</div>

**Algorithme 1**: Lire les pages unes par unes jusqu'à trouver le mot

**Algorithme 2**: Ouvrir le dictionnaire au milieu, se servir d’un mot de la page pour éliminer la moitié du dictionnaire, recommencer la recherche en ouvrant à nouveau l’autre moitié en son milieu, etc.

Le premier algorithme correspond à la **recherce séquentielle**. Le deuxième algorithme correspond à la **recherche dichotomique**.

<div style='color: cadetblue'>
    <h4><u>Problème 2</u> :  Retourner la valeur minimale d'un tableau de <em>n</em> entiers</h4>
</div>

**Algorithme 1**: Parcourir le tableau de gauche à droite en retenant la plus petite valeur

**Algorithme 2**: Trier le tableau dans l’ordre croissant, retourner la première valeur

Dans les deux cas _n_ a exprimé la taille du problème. En fonction de cette taille, la complexité d'un algorithme est le nombre _d'opérations élémentaires_ nécessaires à son éxecution. On appelle _opération élémentaire_ toute opération du type affectation, comparaison, opération arithmétiques.

    <h2>Qu'est-ce que la complexité des algorithmes ?</h2>
</div>

<div style='color: forestgreen'>
    <h3>Notion de complexité temporelle</h3>
</div>

Idéalement, la complexité temporelle d’un algorithme $A$ serait une fonction

$T_{a}(N)$ = Temps nécessaire pour traiter un problème de taille _n_

L’intérêt c'est de choisir entre plusieurs algorithmes pour la même tâche, ou prédire le temps que cela prendra. En réalité le temps de d'exécution d'un algorithme sur machine ne dépend pas seulement de l'algorithme et de _n_. D'autres facteurs peuvent entrer en ligne de compte comme le choix d'implémentation, le compilateur, la machine...

On étudie donc la complexité :

 - en comptant le nombre d’exécutions de certaines instructions, par exemple : "combien de comparaisons ?"
 - en donnant une **classe de complexité** à certains de problèmes qui ressemblent à des problèmes connus dont la complexité est similaire
 - en l'exprimant en fonction de _n_ qui est la taille des données que l'algorithme doit traiter.

   Notations $O$, $\theta$, $\Omega$ 
    
   </h2>

<div style='color: forestgreen'>
    <h3>Intuition</h3>
</div>

Si n est la taille du problème que notre algorithme doit traiter, $f(n)$ est une **estimation du nombre d'opérations nécessaires à l'exécution de l'algorithme en fonction de n**. On s'intéresse au comportement de cette fonction positive quand n tend vers +$\infty$ ce qui nous oblige à introduire de nouvelles notations.

Soient $f$ et $g$ deux fonctions définies selon le formalisme énoncé ci-dessus. On dit :

💡 $f(n) = o(g(n))$ si $f(n)$ **est négligeable devant** $g(n)$ quand n tend vers +$\infty$.

Par exemple $n^{2}$ = $o(n^{3})$ MAIS $\frac {n^{3}} {100}$ $\ne$ o($n^{3}$)

💡 $f(n) = O(g(n))$ si $f(n)$ **est au plus proportionnelle à** $g(n)$ quand n tend vers +$\infty$.

Par exemple $10n^{2}$ = $O(n^{3})$, $\frac {n^{3}} {100}$ = $O(n^{3})$  et on a aussi $100{n^{3}} = O(n^{3})$

💡 $f(n)$ = $\theta(g(n))$ si $f(n)$ **est proportionnelle à ou du même ordre que** $g(n)$ quand n tend vers +$\infty$.

Par exemple  $\frac {n^{3}} {100}$ = $\theta(n^{3})$, $10n^{2} + n$ = $\theta(n^{2})$ MAIS $10n^{2}$ $\ne$ $\theta(n^{3})$

💡 $f(n)$ = $\Omega(g(n))$ si $f(n)$ **est au moins proportionnelle à** $g(n)$ quand n tend vers +$\infty$.

Par exemple  $\frac {n^{3}} {100}$ = $\Omega(n^{2})$, $10n^{2}$ = $\Omega(3n^{2})$ MAIS $10n^{2}$ $\ne$ $\Omega(n^{3})$

Les notations les plus utilisées sont $\theta$ et $O$. Par la suite nous privilegierons donc celles-ci.
     
   </p>
</div>

<div style='color: forestgreen'>
    <h3>Définitions mathématiques</h3>
</div>

**$f(n) \in O(g(n))$** si $\exists c>0, \exists n_{0} > 0$ tel que  $\forall n > n_{0}$ on a $ 0 \leq f(n) \leq cg(n) $

<img src=" " width="250">

**$f(n) \in \theta(g(n))$** si $\exists c_{1}>0, \exists c_{2}>0, \exists n_{0} > 0$ tel que  $\forall n > n_{0}$ on a $ c_{1}g(n) \leq f(n) \leq c_{2}g(n) $

<img src=" " width="250">

**$f(n) \in \Omega(g(n))$** si $\exists c>0, \exists n_{0} > 0$ tel que  $\forall n > n_{0}$ on a $ 0 \leq cg(n) \leq f(n) $

<img src=" " width="250">

    <p style="background-color: #e7f2fa; position: relative; margin: 0 -.6rem !important; padding: .4rem .6rem .4rem 2rem; font-weight: 700;">📝 Note</p>
    <p style="padding: 0 1.4rem; margin-top: .4em; margin-bottom: 0; font-size: 1em;">
        
$O(g(n))$, $\theta(g(n))$ et $\Omega(g(n))$ <b>sont des ensembles</b>, mais on abuse des notations en confondant les ensembles et leurs éléments. 

Ainsi $\theta(n^{2})$ + $\theta(n^{2})$ = $\theta(n^{2})$ <b>abrège l'écriture</b> $\forall f \in \theta(n^{2}), \forall g \in \theta(n^{2}), f + g \in \theta(n^{2})$ 
     
   </p>
</div>

   Utilisation des $O$, $\theta$, $\Omega$ dans les calculs de complexité

<div style='color: forestgreen'>
    <h3>Exemple du tri par insertion</h3>
</div>

On considère l'algorithme suivant :

```python
def selection_sort(array):
    n = len(array)
    for i in range(0, n-1):
        minpos = i
        for j in range(i+1, n):
            if array[j] < array[minpos]:
                minpos = j
        array[i], array[minpos] = array[minpos], array[i]
    return array
```

Peut-on déterminer sa complexité à l'aide des notations asymptotiques ?

<div style='color: cadetblue'>
    <h4>Opérations élémentaires</h4>
</div>

La première ligne de code `n = len(array)` correspond à une opération élémentaire (en l'occurence une affectation). Elle prend un temps constant à la machine ce que nous noterons **O(1)**.

<img src=" " width="500">

<div style='color: cadetblue'>
    <h4>Le cas du <code>for</code></h4>
</div>

Un `for` est une boucle qui va répéter des opérations un nombre de fois déterminé. Sauf accident, ou rupture volontaire de la part de l'utilisateur ou du codeur, le nombre d'opérations est déterminé à l'avance (par les bornes de la fonction range) et ne peut pas changer. On associe donc le for à la notation $\theta$.

Pour savoir quelle grandeur mettre dans theta on regarde les paramètres de **_range_** (ou la taille de la liste si le _range_ n'est pas utilisé). Ici `for i in range(0, n-1)` va de **0** à **n-1** ce qui corespond à n opérations effectives, on met donc un _n_ dans $\theta$ ce qui nous donne $\theta(n)$.

**Notez que toutes les instructions qui sont dans le `for` héritent de la même complexité.**

<img src=" " width="500">

<div style='color: cadetblue'>
    <h4>Le cas du <code>for</code> (2)</h4>
</div>

Lorsque qu'une boucle `for` est comprise dans une autre boucle `for`, elle hérite de la complexité de la boucle `for` dans laquelle elle se trouve à laquelle elle vient ajouter sa propre complexité.

Les bornes du deuxième `for` vont _i+1_ à _n_. Ce for va donc s'effectuer _n-i fois_. Quand _n_ prend une valeur très grande _n-i_ est asymptotiquement équivalent à _n_. Comme il s'agit d'un `for` on met donc $\theta(n)$ mais ce `for` étant compris dans un autre for en $\theta(n)$, **on multiplie leur complexité** pour obtenir $\theta(n^{2})$

<img src=" " width="500">

<div style='color: cadetblue'>
    <h4>Le cas du <code>if</code></h4>
</div>

L'instruction `if` a ceci d'intéressant : Elle n'est pas toujours exécutée, cela dépend des conditions. Comme elle est dans le deuxième `for` elle est censée hériter de sa complexité ($\theta(n^{2})$).

Ceci dit, il faut comprendre que toutes les instructions qui sont comprises dans ce `if` ne seront pas exécutées à chaque tour. Le nombre d'opérations y est donc **au plus** égal à $n^{2}$... Ce qui correspond exactement à la notation O. 
Le code compris dans le `if` est donc en $O(n^{2})$.

<img src=" " width="500">

Au final : $T(n) =  θ(1) + θ(n) + θ(n) + θ(n^{2}) + θ(n^{2}) + O(n^{2}) + θ(n) =  θ(n^{2}).$

Le tri par sélection est un **algorithme quadratique**.
